```python
import pandas as pd 
import seaborn as sns   
```

```python
from sklearn import datasets
```

```python
iris = datasets.load_iris()
```

```python
# from sklearn.datasets import make_classification
# X, y = make_classification(n_samples=1000, n_features=20, n_classes=2)
```

```python
df = pd.DataFrame(data = iris.data, columns= iris.feature_names)
```

```python
# df['target'] = iris.target
# df['target'].unique()
```

```python
y = iris.target
```

```python
X = df.to_numpy()
print(X)
```

```python
import torch 
from torch.utils.data import TensorDataset
```

```python
dataset = TensorDataset(torch.tensor(X),torch.tensor(y))

#access an individual sample
input_sample, label_sample = dataset[0]
print('input_sample:', input_sample)
print('label_sample:', label_sample)
```

Once we create dataset using tensor.Datasets we use data loader for efficient data loading during training  

```python
from torch.utils.data import DataLoader 

batch_size =2      # how many samples are included in each iterations, batching help multiple samples at once
shuffle = True     # help shuffle data after each epoch helping generalization of models
```

epoch: one full pass through the training dataloader 
generalization: model performs well with unseen data.

```python
#create a data loader
dataloader = DataLoader(dataset, batch_size, shuffle)
```

iteration over data loader 

```python
for batch_input, batch_labels in dataloader :   #realtime generally batch_size > 32 for computational efficiency
    print('batch_input', batch_input)
    print('batch_labels', batch_labels)
    
```
